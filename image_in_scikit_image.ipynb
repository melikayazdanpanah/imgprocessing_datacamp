```python
# Import the necessary modules
from skimage import data, color
import matplotlib.pyplot as plt

# Load the rocket image
rocket = data.rocket()

# Convert the image to grayscale
gray_scaled_rocket = color.rgb2gray(rocket)

# Function to display images
def show_image(image, title):
    plt.imshow(image, cmap='gray' if len(image.shape) == 2 else None)
    plt.title(title)
    plt.axis('off')
    plt.show()

# Show the original image
show_image(rocket, 'Original RGB image')

# Show the grayscale image
show_image(gray_scaled_rocket, 'Grayscale image')
```

```python
import numpy as np
# Flip the grayscale image vertically
seville_vertical_flip = np.flipud(gray_scaled_rocket)

# Flip the vertically flipped image horizontally
seville_horizontal_flip = np.fliplr(seville_vertical_flip)

# Show the resulting flipped image
show_image(seville_horizontal_flip, 'Seville')
```

```python
# Load an image using skimage's data module
image = data.rocket()

# Display the original image
plt.figure(figsize=(8, 6))
plt.imshow(image)
plt.title('Original Image')
plt.axis('off')
plt.show()
```

```python
# Check the type of the image object
print(f"Image type: {type(image)}")

# Display image shape (height, width, color channels)
print(f"Image shape: {image.shape}")
```

```python
# Extract Red, Green, and Blue channels
red_channel = image[:, :, 0]
green_channel = image[:, :, 1]
blue_channel = image[:, :, 2]

# Display the original image and each color channel
plt.figure(figsize=(12, 10))

# Original Image
plt.subplot(2, 2, 1)
plt.imshow(image)
plt.title('Original Image')
plt.axis('off')

# Red Channel
plt.subplot(2, 2, 2)
plt.imshow(red_channel, cmap='Reds')
plt.title('Red Channel')
plt.axis('off')

# Green Channel
plt.subplot(2, 2, 3)
plt.imshow(green_channel, cmap='Greens')
plt.title('Green Channel')
plt.axis('off')

# Blue Channel
plt.subplot(2, 2, 4)
plt.imshow(blue_channel, cmap='Blues')
plt.title('Blue Channel')
plt.axis('off')

plt.show()
```

```python
# Convert the image to grayscale
gray_image = color.rgb2gray(image)

# Display the grayscale image
plt.figure(figsize=(8, 6))
plt.imshow(gray_image, cmap='gray')
plt.title('Grayscale Image')
plt.axis('off')
plt.show()
```

```python
# Output the shape and the total number of pixels
image_shape = image.shape
print(f"Image shape: {image_shape}")
total_pixels = image_shape[0] * image_shape[1] * image_shape[2]
print(f"Total number of pixels: {total_pixels}")
```

```python
# Flip the image vertically
vertical_flip = np.flipud(image)

# Display the vertically flipped image
plt.figure(figsize=(8, 6))
plt.imshow(vertical_flip)
plt.title('Vertically Flipped Image')
plt.axis('off')
plt.show()
```

```python
# Flip the image horizontally
horizontal_flip = np.fliplr(image)

# Display the horizontally flipped image
plt.figure(figsize=(8, 6))
plt.imshow(horizontal_flip)
plt.title('Horizontally Flipped Image')
plt.axis('off')
plt.show()
```

```python
# Calculate and plot the histogram for the grayscale image
plt.figure(figsize=(8, 6))
plt.hist(gray_image.ravel(), bins=256, color='black', alpha=0.7)
plt.title('Histogram of Grayscale Image')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')
plt.show()
```

```python
# Create histograms for each color channel
colors = ('red', 'green', 'blue')
plt.figure(figsize=(8, 6))
for i, color_name in enumerate(colors):
    plt.hist(image[:, :, i].ravel(), bins=256, color=color_name, alpha=0.5)
    
plt.title('Color Channel Histograms')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')
plt.legend(colors)
plt.show()
```

**image thresholding**


```python
from skimage import data, color, filters
from skimage.util import img_as_ubyte
```

```python
# Load an example image from scikit-image
image = data.coins()  # You can use other images like data.coins() or data.astronaut()

# Display the original image
plt.figure(figsize=(8, 6))
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')
plt.show()
```

# Thresholding
255(white) if pixel > thresh value

0(black) if pixel < thresh value

Applying Global Thresholding

```python
# Set a threshold value
thresh = 127

# Apply global thresholding
binary_global = image > thresh

# Display the thresholded image
plt.figure(figsize=(8, 6))
plt.imshow(binary_global, cmap='gray')
plt.title('Global Thresholding')
plt.axis('off')
plt.show()
```

Applying Inverted Thresholding

```python
# Apply inverted thresholding
binary_inverted = image <= thresh

# Display the inverted thresholded image
plt.figure(figsize=(8, 6))
plt.imshow(binary_inverted, cmap='gray')
plt.title('Inverted Thresholding')
plt.axis('off')
plt.show()
```

Applying Global Thresholding Algorithms

```python
from skimage.filters import try_all_threshold
import matplotlib.pyplot as plt

image = data.coins()
#obtain all the resulting images
fig , ax = try_all_threshold(image , verbose=False)

#showing resulting plot
plt.show()
```

## Optimal thresh Global Thresholding

for uniform background

```python
from skimage.filters import threshold_otsu

#obtain the optimal threshold value
thresh = threshold_otsu(image)

binary_global = image > thresh

#show
show_image(image , 'original')
show_image(binary_global , 'global thresholding')

```

## Applying Local Thresholding

uneven background

```python
from skimage.filters import threshold_local
#set block size to 35
block_size =35
text_image = data.text()

#obtain the optimal local thresholding 
local_thresh = threshold_local(text_image , block_size , offset =10)
binary_local = text_image > local_thresh
#show
show_image (text_image ,'original')
show_image(binary_local, ' local thresholding')
```
